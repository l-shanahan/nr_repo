# 10th Plot

```python
fig10_1_cut_list = np.load('fig10_1_cut_list.npy')
fig10_1_fppm_list = np.load('fig10_1_fppm_list.npy')

fpr_keras = np.load('fig10_ROC_fpr_keras.npy')
tpr_keras = np.load('fig10_ROC_tpr_keras.npy')
auc = np.load('fig10_ROC_auc.npy')

fig10_2_cut_list = np.load('fig10_2_cut_list.npy')
fig10_2_fppm_list = np.load('fig10_2_fppm_list.npy')

fig10_3_cut_list = np.load('fig10_3_cut_list.npy')
fig10_3_fppm_list = np.load('fig10_3_fppm_list.npy')
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2)

ax1.plot(fig10_1_cut_list,fig10_1_fppm_list)
ax1.set_title("(a)")
ax1.set_ylabel("False Positives / Migdal", fontsize=16, labelpad=5)
ax1.set_xlabel("Cut on 1st CNN", fontsize=16, labelpad=5)

ax2.plot(fpr_keras, tpr_keras, label=f'AUC: {auc}', color='#ee6677')
ax2.legend(loc='lower right')
ax2.set_title("(b)")
ax2.set_ylabel("True Positive Rate", fontsize=16, labelpad=5)
ax2.set_xlabel("False Positive Rate", fontsize=16, labelpad=5)

ax3.plot(fig10_2_cut_list,fig10_2_fppm_list, color='#228833')
ax3.set_title("(c)")
ax3.set_ylabel("False Positives / Migdal", fontsize=16, labelpad=5)
ax3.set_xlabel("Cut on 2nd CNN", fontsize=16, labelpad=5)

ax4.plot(fig10_3_cut_list,fig10_3_fppm_list, color='#aa3377')
ax4.axvline(0.0004,color='#66ccee',linestyle = '--',linewidth=2)
ax4.set_title("(d)")
ax4.set_ylabel("False Positives / Migdal", fontsize=16, labelpad=5)
ax4.set_xlabel("Cut on 2nd CNN", fontsize=16, labelpad=5)

plt.savefig('Fig10');
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,4))

ax1.plot(fig10_1_cut_list,fig10_1_fppm_list)
ax1.axvline(0.006,color='#ee6677',linestyle = '--',linewidth=2,label='Optimal Cut')
ax1.set_title("(a)")
ax1.set_ylabel("False Positives / Migdal", fontsize=16, labelpad=5)
ax1.set_xlabel("Cut on 1st CNN", fontsize=16, labelpad=5)
ax1.legend(loc='lower right', fontsize=16)

ax2.plot(fpr_keras, tpr_keras, label=f'AUC: {auc}', color='#ee6677')
ax2.legend(loc='lower right', fontsize=16)
ax2.set_title("(b)")
ax2.set_ylabel("True Positive Rate", fontsize=16, labelpad=5)
ax2.set_xlabel("False Positive Rate", fontsize=16, labelpad=5)

plt.savefig('Fig10_1');
```
